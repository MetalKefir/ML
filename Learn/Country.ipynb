```python
import pandas as pd
import os

HOUSING_PATH = os.path.join("datasets", "countries-of-the-world")

def load_data(housing_path=HOUSING_PATH):
    csv_path = os.path.join(housing_path, "countries of the world.csv")
    return pd.read_csv(csv_path)
```

```python
country = load_data()
```

```python
import matplotlib.pyplot as plt

country.hist(bins=50, figsize=(20,15))
plt.show()
```
